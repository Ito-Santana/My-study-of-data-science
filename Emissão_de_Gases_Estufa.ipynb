```python
colunas_anos = list(Emissoes_Gases.loc[:,1970:2021].columns)
colunas_anos
```

utilizaremos o metodo melt para transformar colunas em linhas

```python
Emissoes_GasesDF.melt(id_vars = colunas_info , value_vars= colunas_anos, var_name = 'Ano' , value_name='Emissão' )
```

```python
Emissoes_por_ano = Emissoes_GasesDF.melt(id_vars = colunas_info , value_vars= colunas_anos, var_name = 'Ano' , value_name='Emissão' )
```

```python
Emissoes_por_ano.groupby('Gás')
```

```python
Emissoes_por_ano.groupby('Gás').groups
```

```python
Emissoes_por_ano.groupby('Gás').get_group('CO2 (t)')
```

```python
Emissoes_por_ano.groupby('Gás').sum(numeric_only = True).sort_values('Emissão', ascending = False)
```
