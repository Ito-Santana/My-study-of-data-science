#Importando os dados

```python
import pandas as pd

url='https://caelum-online-public.s3.amazonaws.com/2928-transformacao-manipulacao-dados/dados_locacao_imoveis.json'

dados = pd.read_json(url)
dados
```

Agora devemos normalizar a coluna dados_locacao

```python
dados= pd.json_normalize(dados['dados_locacao'])
dados
```

#Etapa 2

o desafio do projeto 2 está listado em algumas metas:

* Remover os dados em listas dentro do DataFrame;
* Verificar os tipos de dados;
* Identificar colunas numéricas;
* Transformar a coluna numérica para o tipo numérico.

```python
dados
```

percebemos que a apenas a 1 coluna não é do tipo lista

```python
colunas = list(dados.columns)
colunas
```

```python
dados = dados.explode(colunas[1:])
dados
```

agora nosso dado esta em formato de string ou numerico, mas para confirmar devemos usar .info()

```python
dados.info()
```

O correto seria que nosso coluna 'Valor_alguel' fosse do tipo float para realizarmos operações numericas

```python
dados['valor_aluguel'] = dados['valor_aluguel'].apply(lambda x : x.replace('$', '').replace(',','.').replace('reais' , '').strip() )
dados
```

Agora removemos os seguintes termos " '$' ,   ','   , 'reais'" deixando tudo no padrão de numero


```python
dados['valor_aluguel'] = dados['valor_aluguel'].astype(float)
dados.info()
```

Agora transformamos nossa coluna no tipo float

#ETAPA 3

Buscando explicar a organização da identificação dos apartamentos, durante a criação do conjunto de dados, foi adicionado o texto (blocoAP). Esse texto informa que os nomes dos apartamentos estão organizados com a letra do bloco seguida do número do apartamento. No entanto, isso não traz nenhuma informação para nossos dados, sendo interessante realizar a remoção desse texto no conjunto de dados.

Com isso, manipule os textos na coluna apartamento para remover o texto (blocoAP) do DataFrame.

```python
dados.head()
```

```python
dados['apartamento'] = dados['apartamento'].str.replace('\([^)]*\)' , '', regex=True)
dados['apartamento']
```
